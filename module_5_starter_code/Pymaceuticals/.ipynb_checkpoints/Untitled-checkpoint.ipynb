```python
#Dpendencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
Combined_Mousedf = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
Combined_Mousedf.head(30)
```

```python
Mouse_count = Combined_Mousedf["Mouse ID"].unique()
Mouse_count = len(Mouse_count)
Mouse_count
```

```python
#Find the dup Mouse ID
Duplicate = Combined_Mousedf[len("Mouse ID") > 9 ] 
Duplicate
```

```python
#don't drop the duplicate, drop all data from that mouse ID

CleanedDF = Combined_Mousedf.drop(
subset = ['Mouse ID', 'Timepoint'],
keep = 'last').reset_index(drop = True)
CleanedDF


# Cleadeddftest = Cleadeddftest.Index.drop_duplicates("Mouse ID", keep='first')
```
